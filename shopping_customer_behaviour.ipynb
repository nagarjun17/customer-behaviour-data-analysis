```python
import pandas as pd
#load data into dataframe
df= pd.read_csv('customer_shopping_behaviour.csv')
```

```python
#by default gives first 5 values
df.head()
```

```python
df.info()
```

```python
#gives all descriptive statistics of the DataFrame
df.describe(include='all')
```

```python
df.isnull().sum() #to check for any missing values and the number of count
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
#For each group, calculate the median and fill missing values with that median, keeping the original DataFrame size.
```

```python
df.isnull().sum()
```

```python
#to change all the column name to snake case
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
#rename columns heading
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#To create new column for age group
labels = ['Young Adult','Adult','Middle Aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#Create column for purchase_frequency_days
frequency_mapping = {
    'Weekly': 7,
    'Fortnightly': 14,
    'Monthly': 30,
    'Quarterly': 90,
    'Halferly': 180,
    'Annually': 365,
    'Bi-Weekely': 14,
    'Every Three Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping).fillna(0).astype(int) #change the float value to integer
#fillna(0) replace nan value to 0
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```
